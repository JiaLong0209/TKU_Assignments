```python
class EmbeddedSystem:
    def __init__(self, name, students: dict = None):
        self.name = name
        self.students = students if students is not None else {}

    def display_name(self):
        print(f"Embedded System Name: {self.name}")

    def update_name(self, new_name):
        self.name = new_name

    def print_students(self):
        for id, student in self.students.items():
            print(f"ID: {id}, Name: {student}")

    def update_student(self, id, name):
        self.students[id] = name

    def get_student(self, id):
        return self.students.get(id, "Student not found")

    def get_stu_id(self, name):
        for id, student in self.students.items():
            if student == name:
                return id
    

stduents = {
    "1101": "Alice",
    "1102": "Bob",
    "1103": "Charlie"
}

es = EmbeddedSystem("TKU Embedded System", stduents)
es.display_name()
es.update_name("New TKU Embedded System")
es.print_students()
es.update_student("1104", "Jialong")
es.display_name()
es.print_students()

print(es.get_student("1102"))
print(es.get_stu_id("Charlie"))

```

```python
import serial 
import time

ser = serial.Serial('/dev/ttyACM1', 9600)

if ser.is_open:
    print("Port is open")
    count = 0

    try: 
        while True:
            if ser.in_waiting > 0:
                data = ser.readline().decode('utf-8').strip()
                print(f"[{count}] Received:", data)
                count += 1
            time.sleep(0.1)

    except KeyboardInterrupt:
        print("Exiting program.")

    finally:
        ser.close()
        print("Port is closed")

else:
    print("Port is not open")

            
    
    
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QPushButton, QVBoxLayout, QLineEdit

class ProjectApp(QWidget):
    def __init__(self):
        super().__init__() # Call the constructor of the parent class (QWidget)
        self.initalizeUI()

    def initalizeUI(self):
        # Set up the window
        self.setWindowTitle('Basic PyQt UI')
        self.setGeometry(300, 300, 300, 200) # x, y, width, height
        # Layout
        layout = QVBoxLayout()
        # Widgets
        self.label = QLabel('Enter your name: ', self)
        self.textbox = QLineEdit()
        self.button = QPushButton('Submit')
        self.button.clicked.connect(self.handleButtonClick)

        # Add a button
        layout.addWidget(self.label)
        layout.addWidget(self.textbox)
        layout.addWidget(self.button)
        self.setLayout(layout)
        self.show()

    def handleButtonClick(self):
        name = self.textbox.text()
        self.label.setText(f'Hello, {name}')

if __name__ == '__main__':
    app = QApplication(sys.argv)
    appWindow = ProjectApp()
    sys.exit(app.exec_()) # app.exec_() starts the application’s event loop
```

```python

import sys
import serial
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QVBoxLayout
# Initialize serial communication
ser = serial.Serial('/dev/ttyACM1', 9600)

class ArduinoApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Arduino Controller')
        self.setGeometry(300, 300, 300, 150)
        # Button to turn LED on
        self.on_button = QPushButton('Turn LED On')
        self.on_button.clicked.connect(self.turn_on_led)
        # Button to turn LED off
        self.off_button = QPushButton('Turn LED Off')
        self.off_button.clicked.connect(self.turn_off_led)
        # Layout
        layout = QVBoxLayout()
        layout.addWidget(self.on_button)
        layout.addWidget(self.off_button)
        self.setLayout(layout)
        self.show()

    def turn_on_led(self):
        ser.write(b'1') # Send '1' to Arduino

    def turn_off_led(self):
        ser.write(b'0') # Send '0' to Arduino

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = ArduinoApp()
    sys.exit(app.exec_())
```

```python
import sys
import serial
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QVBoxLayout
from PyQt5.QtCore import QTimer
ser = serial.Serial('/dev/ttyACM0', 9600)
# ser = serial.Serial('COM3', 9600)

class SensorMonitor(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Monitoring System')
        self.setGeometry(300, 300, 300, 150)
        self.label = QLabel('Potentiometer level: ', self)
        layout = QVBoxLayout()
        layout.addWidget(self.label)
        self.setLayout(layout)
        # Timer to update the temperature every second
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.updateSensorData)
        self.timer.start(1000)
        self.show()

    def updateSensorData(self):
        if ser.in_waiting > 0:
            readData = ser.readline().decode('utf-8').strip()
            self.label.setText(f'Potentiometer level: {readData}')

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = SensorMonitor()
    sys.exit(app.exec_())
```
